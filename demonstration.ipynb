визуализация - в этом ноутбуке


в этом коде использовались заранее написанные полиномы лежандра и чебышева. можно брать свои, для этого создайте обьект класса ODE и задайте коэффициенты, а также граничные условия.


```python
import matplotlib.pyplot as plt
from problems.legandre_problem import Legandre
from problems.chebyshev_problem import Chebyshev
from solver.Progonka_solver import Progonka
from problems.lin_generator import ODE


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
ax1.set_title("Полиномы Лежандра")
ax1.set_xlabel("x")
ax1.set_ylabel("y(x)")
ax2.set_title("Полиномы Чебышева")
ax2.set_xlabel("x")
ax2.set_ylabel("y(x)")

for bmn in range(5):
    tak1 = Legandre(bmn)
    A1, b1 = tak1.generate_system(100)
    solver1 = Progonka()
    solver1.solve(A1, b1)
    y_1 = solver1.get_solution()
    x_1 = tak1.x_grid
    ax1.plot(x_1, y_1, lw=1.2, label=f"n = {bmn}", zorder=3)

    tak2 = Chebyshev(bmn)
    A2, b2 = tak2.generate_system(100)
    solver2 = Progonka()
    solver2.solve(A2, b2)
    y_2 = solver2.get_solution()
    x_2 = tak2.x_grid
    ax2.plot(x_2, y_2, lw=1.2, label=f"n = {bmn}", zorder=3)

ax1.grid(alpha=0.5)
ax1.legend(fontsize=10)
ax2.grid(alpha=0.5)
ax2.legend(fontsize=10)
plt.tight_layout()
plt.show()
```
